# Pandas API w Spark (pandas API on Spark / Koalas)

Pandas API on Spark (wcześniej znane jako Koalas) to interfejs w PySpark, który pozwala używać znanej składni Pandas, ale działa rozproszonym silnikiem Spark.

> „Piszesz jak w Pandas, działa jak w Spark!”

Od wersji **spark 3.2**

https://spark.apache.org/pandas-on-spark/

````python
import pyspark.pandas as ps
import os
os.environ["PYARROW_IGNORE_TIMEZONE"] = "1"
````

````python
import pandas as pd

pdf = pd.DataFrame({
    "id": [1, 2, 3],
    "value": [10, 20, 30]
})
pdf["value_squared"] = pdf["value"] ** 2
````

````python
import pyspark.pandas as ps

psdf = ps.DataFrame({
    "id": [1, 2, 3],
    "value": [10, 20, 30]
})
psdf["value_squared"] = psdf["value"] ** 2
````

| Cecha        | Pandas                   | Pandas API on Spark             |
| ------------ | ------------------------ | ------------------------------- |
| Silnik       | Pamięć (RAM, lokalnie)   | Spark (rozproszony, YARN, itd.) |
| Skala danych | Mała (< RAM)             | Duża (klaster)                  |
| Wydajność    | Szybka dla małych danych | Lepsza dla dużych danych        |
| API          | pandas                   | prawie 1:1 kompatybilne         |


| Narzędzie           | Dla kogo?                        |
| ------------------- | -------------------------------- |
| Pandas              | Analitycy, małe zbiory danych    |
| PySpark SQL         | Programiści, duże zbiory danych  |
| Pandas API on Spark | Analitycy przechodzący do Sparka |
